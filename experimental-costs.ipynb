```python
def plot_fractions(service_costs, paid_costs, total_costs, dataset, path_to_save, colorful):
    service_cost_frac = list( list(map(operator.truediv, service_cost, total_cost)) for service_cost, total_cost in zip(service_costs, total_costs))
    paid_cost_frac = list( list(map(operator.truediv, paid_cost, total_cost)) for paid_cost, total_cost in zip(paid_costs, total_costs))

    counter_determ_service = [item[0] for item in service_cost_frac]
    counter_non_determ_service = [item[1] for item in service_cost_frac]
    random_reset_service = [item[2] for item in service_cost_frac]
    timestamp_service = [item[3] for item in service_cost_frac]

    counter_determ_paid = [item[0] for item in paid_cost_frac]
    counter_non_determ_paid = [item[1] for item in paid_cost_frac]
    random_reset_paid = [item[2] for item in paid_cost_frac]
    timestamp_paid = [item[3] for item in paid_cost_frac]
    
    costs = [counter_determ_service, counter_determ_paid, counter_non_determ_service, counter_non_determ_paid, \
            random_reset_service, random_reset_paid, timestamp_service, timestamp_paid]


    ind = np.arange(len(counter_determ_service))
    width = 0.5

    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,10))
    
    patches_handle = []

    if (colorful):
        patches_handle.append(axes[0][0].bar(ind, counter_determ_service, width, label="Service cost"))
        patches_handle.append(axes[0][0].bar(ind, counter_determ_paid, width, bottom=counter_determ_service, label="Paid cost"))

        patches_handle.append(axes[0][1].bar(ind, counter_non_determ_service, width, label="Service cost"))
        patches_handle.append(axes[0][1].bar(ind, counter_non_determ_paid, width, bottom=counter_non_determ_service, label="Paid cost"))

        patches_handle.append(axes[1][0].bar(ind, random_reset_service, width, label="Service cost"))
        patches_handle.append(axes[1][0].bar(ind, random_reset_paid, width, bottom=random_reset_service, label="Paid cost"))

        patches_handle.append(axes[1][1].bar(ind, timestamp_service, width, label="Service cost"))
        patches_handle.append(axes[1][1].bar(ind, timestamp_paid, width, bottom=timestamp_service, label="Paid cost"))
    else:
        patches_handle.append(axes[0][0].bar(ind, counter_determ_service, width, label="Service cost", color = 'dimgray'))
        patches_handle.append(axes[0][0].bar(ind, counter_determ_paid, width, bottom=counter_determ_service, label="Paid cost", color = 'silver'))

        patches_handle.append(axes[0][1].bar(ind, counter_non_determ_service, width, label="Service cost", color = 'dimgray'))
        patches_handle.append(axes[0][1].bar(ind, counter_non_determ_paid, width, bottom=counter_non_determ_service, label="Paid cost", color = 'silver'))
        
        patches_handle.append(axes[1][0].bar(ind, random_reset_service, width, label="Service cost", color = 'dimgray'))
        patches_handle.append(axes[1][0].bar(ind, random_reset_paid, width, bottom=random_reset_service, label="Paid cost", color = 'silver'))

        patches_handle.append(axes[1][1].bar(ind, timestamp_service, width, label="Service cost", color = 'dimgray'))
        patches_handle.append(axes[1][1].bar(ind, timestamp_paid, width, bottom=timestamp_service, label="Paid cost", color = 'silver'))

    axes[0][0].set_title("Counter deterministic")
    axes[0][1].set_title("Counter non-deterministic")
    axes[1][0].set_title("Random reset")
    axes[1][1].set_title("Timestamp")
    
    for j in range(len(patches_handle)):
        for i, patch in enumerate(patches_handle[j].get_children()):
            bl = patch.get_xy()
            x = 0.5*patch.get_width() + bl[0]
            y = 0.5*patch.get_height() + bl[1]
            if (j == 0 or j == 1):
                axes[0][0].text(x,y, "%d%%" % (int(round(costs[j][i], 2) * 100)), ha='center')
            if (j == 2 or j == 3):
                axes[0][1].text(x,y, "%d%%" % (int(round(costs[j][i], 2) * 100)), ha='center')
            if (j == 4 or j == 5):
                axes[1][0].text(x,y, "%d%%" % (int(round(costs[j][i], 2) * 100)), ha='center')
            if (j == 6 or j == 7):
                axes[1][1].text(x,y, "%d%%" % (int(round(costs[j][i], 2) * 100)), ha='center')

    xticklabels = ("1", "2", "3", "4", "5", "6", "10", "20", "50", "100")

    for axe in axes:
        for ax in axe:
            ax.set_xticks(ind)
            ax.set_xticklabels(xticklabels)
            ax.legend()
            ax.set_xlabel("d")

#     fig.suptitle(dataset + ": Service-paid cost fractions", fontsize=16)
    # fig.tight_layout()
    fig.subplots_adjust(top=0.92)
    plt.savefig(path_to_save)
    plt.show()
```

```python
def find_all_costs(all_matches_dataset):  
    all_costs = []
    for all_matches_service in all_matches_dataset:
        average_results = []
        for match in all_matches_service:
            match = np.reshape(match, (-1, 4))
            #without kennedy
#             match = match[[0, 1, 2, 3, 4, 6, 7, 8, 9, 10]]
            counter_determ_avg = np.average(np.array([item[0] for item in match]).astype(np.int))
            counter_non_determ_avg = np.average(np.array([item[1] for item in match]).astype(np.int))
            random_reset_avg = np.average(np.array([item[2] for item in match]).astype(np.int))
            timestamp_avg = np.average(np.array([item[3] for item in match]).astype(np.int))


            all_algos = np.array([counter_determ_avg, counter_non_determ_avg, random_reset_avg, timestamp_avg])
            average_results.append(all_algos)  
        all_costs.append(average_results)
    return all_costs 
```

```python
def find_costs(all_costs):
    costs = []
    for d in range(0, 10):
        d_list = [item[d] for item in all_costs]
        averages = []
        for i in range(0, 4):
            average = np.average([item[i] for item in d_list])
            averages.append(average)
        costs.append(averages)
    return costs
```

```python
def find_all_matches_dataset(dataset_path, file_path_prefix):
    files_list = []
    for i in range(1, 11):
        file_path = dataset_path + file_path_prefix + str(i) + ".txt" 
        file = open(file_path, 'r').read()
        files_list.append(file)
        
    ds = [1, 2, 3, 4, 5, 6, 10, 20, 50, 100]
    
    
    #Find all matches for each d for both service and paid
    all_matches_service_dataset = []
    all_matches_paid_dataset = []

    for file in files_list:
        all_matches_service = []
        all_matches_paid = []
        for d in ds:
            regex = str(d) + "\t\d*\t(\d*)\t\d*"
            matchobj = re.findall(regex, file, flags=0)
            all_matches_service.append(matchobj)

            regex = str(d) + "\t\d*\t\d*\t(\d*)"
            matchobj = re.findall(regex, file, flags=0)
            all_matches_paid.append(matchobj)

        all_matches_service_dataset.append(all_matches_service)
        all_matches_paid_dataset.append(all_matches_paid)
    return all_matches_service_dataset, all_matches_paid_dataset
```

```python
def find_averages(dataset_path, file_path_prefix):    
   
    #Take the average service cost over one file
    #all_matches contains 1 array for every d
    #each array contains 44 elements - every four elements belong to a different file.
    #elements one to four belong to the four different algorithms

    all_matches_service_dataset, all_matches_paid_dataset = find_all_matches_dataset(dataset_path, file_path_prefix)
#     print(all_matches_service_dataset)
    all_service_costs = find_all_costs(all_matches_service_dataset)
    all_paid_costs = find_all_costs(all_matches_paid_dataset)

    
    service_costs = find_costs(all_service_costs)
    paid_costs = find_costs(all_paid_costs)
        
    return service_costs, paid_costs
```

```python
def prep_box_data(dataset_path, file_path):
    all_matches_service_dataset, all_matches_paid_dataset = find_all_matches_dataset(dataset_path, file_path)
    all_service_costs = find_all_costs(all_matches_service_dataset)
    all_paid_costs = find_all_costs(all_matches_paid_dataset)

    all_box_data_service = []
    all_box_data_paid = []
    
    for d in range(0, len(all_service_costs[0])):
        algos_values = []
        for algo in range(0, len(all_service_costs[0][0])):
            algo_values = []
            for iteration in range(0, len(all_service_costs)):
                algo_values.append(all_service_costs[iteration][d][algo])
            algos_values.append(algo_values)
        all_box_data_service.append(algos_values)
        
    for d in range(0, len(all_paid_costs[0])):
        algos_values = []
        for algo in range(0, len(all_paid_costs[0][0])):
            algo_values = []
            for iteration in range(0, len(all_paid_costs)):
                algo_values.append(all_paid_costs[iteration][d][algo])
            algos_values.append(algo_values)
        all_box_data_paid.append(algos_values)
        
    return all_box_data_service, all_box_data_paid
                
```

```python
def print_costs_per_file(dataset_path, file_path_prefix, dataset_file_names):
    files_list = []
    for i in range(1, 3):
        file_path = dataset_path + file_path_prefix + "_" + str(i) + ".txt" 
        file = open(file_path, 'r').read()
        files_list.append(file)
        
    ds = [1, 2, 3, 4, 5, 6, 10, 20, 50, 100]
    
    
    #Find all matches for each d for both service and paid
    all_matches_service_dataset = []
    all_matches_paid_dataset = []
    
    for d in ds:
        all_matches_service_d = []
        all_matches_paid_d = []
        for file_name in dataset_file_names:
            all_matches_service_file = []
            all_matches_paid_file = []
            for file in files_list:
                regex ="FILE: " + file_name + "\nALGORITHM: [A-z ]*\nd\tk\tService cost\tPaid exchange cost\n" + str(d) + "\t\d*\t(\d*)\t\d*"
                matchobj = re.findall(regex, file, flags=0)
                all_matches_service_file.append(matchobj)

                regex ="FILE: " + file_name + "\nALGORITHM: [A-z ]*\nd\tk\tService cost\tPaid exchange cost\n" + str(d) + "\t\d*\t\d*\t(\d*)"
                matchobj = re.findall(regex, file, flags=0)
                all_matches_paid_file.append(matchobj)

            all_matches_service_file = list(map(list, zip(*all_matches_service_file)))
            all_matches_paid_file = list(map(list, zip(*all_matches_paid_file)))
            all_matches_service_d.append(all_matches_service_file)
            all_matches_paid_d.append(all_matches_paid_file)
                
        all_matches_service_dataset.append(all_matches_service_d)
        all_matches_paid_dataset.append(all_matches_paid_d)
        
    print(len(all_matches_service_dataset))
    print(len(all_matches_service_dataset[0]))
    print(len(all_matches_service_dataset[0][0]))
    print(len(all_matches_service_dataset[0][0][0]))
    print(all_matches_service_dataset)
    
```

```python
def plot_box_plots(service_costs, dataset, path_to_save, colorful):
    
    fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(20,15))
#     fig.suptitle(dataset, fontsize=16)
    ds = [1, 2, 3, 4, 5, 6, 10, 20, 50, 100]
    
    for d in range(0, len(service_costs)):
        box_plot_data=[service_costs[d][1], service_costs[d][2], service_costs[d][3]]
        row_offset = 0
        col_offset = d
        if (d < 4):
            row_offset = 0
        elif (d < 8):
            row_offset = 1
        else:
            row_offset = 2
            col_offset += 1
        bp = axes[0 + row_offset][col_offset % 4].boxplot(box_plot_data, labels=['CN', 'RR', 'TS'])
        
        if (colorful == False):
            for median in bp['medians']:
                median.set(color='black')
        axes[0 + row_offset][col_offset % 4].set_xlabel("d = " + str(ds[d]))
        axes[0 + row_offset][col_offset % 4].ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    
    axes[2, 0].axis('off')
    axes[2, 3].axis('off')
    
    plt.tight_layout()
    fig.subplots_adjust(top=0.83)
    plt.savefig(path_to_save)
    plt.show()
```
